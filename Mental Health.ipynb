## FREQUENCIES AND VARIABLE DESCRIPTIONS

* CASEID: Case identification number
* YEAR: Reporting period
* AGE: Age (recoded) :: -9 - Missing/Not-Recorded
* EDUC: Education
* SPHSERVICE: State psychiatric hospital services
* CMPSERVICE: SMHA-funded/operated community-based program
* OPISERVICE: Other psychiatric inpatient
* RTCSERVICE: Residential treatment center
* IJSSERVICE: Institutions under the justice system
* MH1: Mental health diagnosis one
* MH2: Mental health diagnosis two
* MH3: Mental health diagnosis three
* SUB: Substance use diagnosis
* MARSTAT: Marital status
* SMISED: SMI/SED status
* SAP: Substance use problem
* EMPLOY: Competitive employment status (aged 16 years and older) at discharge or end of the reporting period
* DETNLF: Detailed 'not in labor force' category
* VETERAN: Veteran status
* LIVARAG: Residential status - at discharge or end of reporting period
* NUMMHS: Number of mental health diagnoses reported
* TRAUSTREFLG: Trauma- or stressor-related disorder reported
* ANXIETYFLG: Anxiety disorder reported
* ADHDFLG: Attention deficit/hyperactivity disorder reported
* CONDUCTFLG: Conduct disorder reported
* DELIRDEMFLG: Delirium/dementia disorder reported
* BIPOLARFLG: Bipolar disorder reported
* DEPRESSFLG: Depressive disorder reported
* ODDFLG: Oppositional defiant disorder reported
* PDDFLG: Pervasive developmental disorder reported
* PERSONFLG: Personality disorder reported
* SCHIZOFLG: Schizophrenia or other psychotic disorder reported
* ALCSUBFLG: Alcohol or substance-related disorder reported
* OTHERDISFLG: Other mental disorder reported
* STATEFIP: Reporting state code
* DIVISION: Census division
* REGION: Census region

## Decoding Anxiety:

The purpose of this part of study would be to identify factors that can help assess the risk for Anxiety Disorder for a particular individual

```python
#Step -1: Dropping features other than anxiety

drop_list_anxiety = ['SPHSERVICE', 'CMPSERVICE','OPISERVICE', 'RTCSERVICE', 
                     'IJSSERVICE', 'MH1', 'MH2', 'MH3','TRAUSTREFLG', 'ADHDFLG', 
                     'CONDUCTFLG', 'DELIRDEMFLG', 'BIPOLARFLG', 'DEPRESSFLG', 
                     'ODDFLG', 'PDDFLG','PERSONFLG', 'SCHIZOFLG', 'ALCSUBFLG', 'OTHERDISFLG', 'DETNLF', 'VETERAN', 'SUB']

data_anxiety = df.drop(drop_list_anxiety, axis=1)
data_anxiety
```

```python
#replace -9 with Null values first then apply imputer
import numpy as np

data_anxiety = data_anxiety.replace(-9.0, np.nan)
data_anxiety
```

```python
data_anxiety.dropna(inplace=True)
data_anxiety.shape
```

```python
from sklearn.model_selection import train_test_split

X = data_anxiety.drop('ANXIETYFLG', axis=1)
y = data_anxiety['ANXIETYFLG']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=30)
X_train.shape, y_train.shape, X_test.shape, y_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt

model_anxiety = RandomForestClassifier(n_estimators=100, random_state=42)
model_anxiety.fit(X_train, y_train)
y_pred = model_anxiety.predict(X_test)

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_pred, labels=model_anxiety.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model_anxiety.classes_)
disp.plot()
plt.show()
```

```python
from sklearn.svm import SVC

anxiety_svc = SVC(random_state=42)
anxiety_svc.fit(X_train, y_train)
y_pred_svc = anxiety_svc.predict(X_test)

cm_svc = confusion_matrix(y_test, y_pred_svc, labels=anxiety_svc.classes_)
disp_svc = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=anxiety_svc.classes_)
disp_svc.plot()
plt.show()
```
